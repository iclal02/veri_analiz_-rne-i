#### tüm sütunlarımız sağa çarpık çıktı genel bir bakışla daha detaylı incelicez

```python
# şimdi aykırı değerleri bulucaz IQR kullanarak
```

```python
import pandas as pd
import numpy as np

# Sayısal sütunları seçtik
numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns
df['aykiri_mi'] = 'Normal'
df['aykiri_sayisi'] = 0
# Eşik değeri belirleyelim (örneğin, 2 sütundan fazla aykırı ise aykırı olarak sınıflandıralım)
esik_deger = 3

for column in numeric_cols:
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Aykırı değerleri belirleyelim ve 'aykiri_sayisi' sütununu artıralım
    df.loc[(df[column] < lower_bound) | (df[column] > upper_bound), 'aykiri_sayisi'] += 1

    # 'aykiri_sayisi' sütunu 2'yi geçen satırları 'aykiri_mi' olarak işaretleyelim
    df.loc[df['aykiri_sayisi'] >= esik_deger, 'aykiri_mi'] = 'Aykırı'

    # Aykırı değerleri sütunlar bazında incelemek için ek bilgi çıktısı
for column in numeric_cols:
    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
    print(f"Aykırı değerler {column} sütununda:")
    print(outliers)
```

```python
df_temiz = df[df['aykiri_mi'] != 'Aykırı'].copy()
df_temiz
```

```python
# şimdi object tanımlanmış sütunlarımızın temizliğini yapıcaz
```

```python
kelimeler=df_temiz.select_dtypes(include=['object']).columns

df_temiz[kelimeler] = df_temiz[kelimeler].apply(lambda x: x.str.capitalize())
df_temiz[kelimeler]
```

```python
for kelime in kelimeler:
    print(f"{kelime} sütunu")
    print(df_temiz[kelime].unique())
    print(df_temiz[kelime].value_counts())
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
klm=['mainroad',
       'guestroom', 'basement', 'hotwaterheating', 'airconditioning','prefarea']
for kelime in klm:
    le = LabelEncoder()
    df_temiz[kelime] = le.fit_transform(df_temiz[kelime])
df_temiz.head()
```

```python
# One-Hot Encoding ile sayısallaştırma
df_temiz = pd.get_dummies(df_temiz, columns=['furnishingstatus'], prefix='furnishing',dtype=int)
df_temiz.head()
```

```python
df_temiz['aykiri_mi'] = df_temiz['aykiri_mi'].replace({'Aykırı': 1, 'Normal': 0})
df_temiz.head()
```

```python
df_temiz['total_rooms'] = df_temiz['bedrooms'] + df_temiz['bathrooms'] + df_temiz['guestroom']
df_temiz
```

```python
df_temiz['price_per_area'] = df_temiz['price'] / df_temiz['area']
df_temiz
```

```python
# Korelasyon matrisini hesapladık
correlation_matrix = df_temiz[["price","area","stories","mainroad","total_rooms","furnishing_Furnished","furnishing_Semi-furnished","furnishing_Unfurnished"]].corr()

# Isı haritasını oluşturduk
plt.figure(figsize=(10, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='magma', fmt='.2f')
plt.title('Özellikler Arasındaki Korelasyon Isı Haritası')
plt.show()
```

```python
import pandas as pd
import numpy as np
from scipy import stats

results=[]

# Her sayısal sütun için ortalama, medyan, mod ve standart sapmayı hesaplama
for column in df_temiz.select_dtypes(include=[np.number]).columns:
    mean_value = df_temiz[column].mean()
    median_value = df_temiz[column].median()
    mode_value = df_temiz[column].mode()[0]  # Modu alırken tek bir değer döndüğünden emin oluyoruz
    std_value = df_temiz[column].std()


    results.append({
        'Sütun': column,
        'Ortalama': mean_value,
        'Medyan': median_value,
        'Mod': mode_value,
        'Standart Sapma': std_value
    })

# Sonuçları DataFrame'e dönüştürme
results_df = pd.DataFrame(results)

results_df
```

```python
#çarpıklık bir verinin ortalamadan ne kadar uzaklaştığını verir
# basıklık verilerin ortalama etrafında nasıl yoğunlaştığını gösterir
```

```python
carpiklik_price = df_temiz['price'].skew()
basiklik_price = df_temiz['price'].kurt()

carpiklik_area = df_temiz['area'].skew()
basiklik_area = df_temiz['area'].kurt()

print("Price sütununun çarpıklığı:", carpiklik_price)
print("Price sütununun basıklığı:", basiklik_price)
print("Area sütununun çarpıklığı:", carpiklik_area)
print("Area sütununun basıklığı:", basiklik_area)
```

```python
#price çarpıklığı: Pozitif bir çarpıklık değeri, fiyatların çoğunlukla daha düşük seviyelerde olduğu, ancak birkaç yüksek fiyatlı gözlemin de olduğu anlamına gelir.
#Yani, fiyat dağılımı sağa doğru uzamış bir şekle sahiptir.
#Bu, fiyatların uzun kuyruğunun sağ tarafta olduğunu gösterir.
```

```python
#price basıklığı :Basıklık değeri normal dağılımın basıklık değerinden (yaklaşık 3) biraz daha düşüktür.
#Bu, fiyat verilerinin normal dağılıma göre biraz daha düz tepeli olduğunu gösterir.
#Yani, fiyatlar ortalama değerin etrafında normal dağılımdan daha yaygın bir şekilde dağılmıştır.
```

```python
# Price sütununda olduğu gibi, area sütununda da pozitif bir çarpıklık değeri vardır.
#Bu, alanların çoğunlukla daha küçük olduğu ve birkaç büyük alanlı gözlemin olduğu anlamına gelir.
#Yani, alan dağılımı da sağa doğru uzamış bir şekle sahiptir.
```

```python
#Area sütununun basıklık değeri, price sütununa göre biraz daha yüksektir. 
#Bu, alan verilerinin normal dağılıma göre biraz daha sivri tepeli olduğunu gösterir.
#Yani, alanlar ortalama değerin etrafında daha sıkı toplanmıştır.
```
